```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset["DESCR"])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(dataset.data,columns=["age",
                                     "sex",
                                     "bmi",
                                     "bp",
                                     "s1",
                                     "s2",
                                     "s3",
                                     "s4",
                                     "s5",
                                     "s6"])
```

```python
df
```

```python
## Independent and dependent Features
X=df
Y=dataset["target"]
```

```python
## Train Test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(X_train.corr())
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,Y_train)
```

## Hyperparameter tuning

```python
params={
    "criterion":["squared_error","friedman_mse","absolute_error"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5,10,15,20,25],
    "max_features":["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring="neg_mean_squared_error")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
Y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(Y_test,Y_pred))
print(mean_squared_error(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion="squared_error",max_depth=3,max_features="auto",splitter="best")
```

```python
selectedmodel.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(selectedmodel,filled=True)
```
